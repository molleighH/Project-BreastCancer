```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic


df = pd.read_csv(r'c:\Users\Sunny\OneDrive\Desktop\Project1\output.csv')

grouped = df.groupby(['Menopause', 'Class']).size().unstack()

#creating chart
grouped.plot(kind='bar', stacked=True)

#adding labels
plt.xlabel('Menopause Status')
plt.ylabel('Count')
plt.title('Correlation between Menopause Status and Recurrence')

# Save chart
plt.savefig('bar_chart.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
df['Menopause'] = df['Menopause'].astype('category')
df['Class'] = df['Class'].astype('category')


contingency_table = pd.crosstab(df['Menopause'], df['Class'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected contingency table:")
print(expected)
```
